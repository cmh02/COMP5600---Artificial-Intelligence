```python
'''
Problem 1

Description: Create a Numpy array of shape [1x5] with random values drawn from a uniform distribution and perform several computations on it.
'''

# Module Imports
import numpy as np

# Create [1x5] Numpy Array with random values
fiveRandomArray = np.random.rand(5)
print(f"The array is: \n{fiveRandomArray}\n")

# Calculate mean and standard deviation
fiveRandomMean = np.mean(fiveRandomArray)
fiveRandomStdDev = np.std(fiveRandomArray)
print(f"The array has a mean of {fiveRandomMean} and a standard deviation of {fiveRandomStdDev}!\n")

# Reshape array to [5x1] 2D array
fiveRandomArray = np.reshape(fiveRandomArray, (5, 1))
print(f"The reshaped 2D array is: \n{fiveRandomArray}\n")

# Add 5 to each element in the array
fiveRandomArray = np.add(fiveRandomArray, 5)
print(f"The array after adding 5 to each element is: \n{fiveRandomArray}\n")

# Compute the dot product of the 2D array with itself
fiveRandomDotProduct = np.dot(fiveRandomArray, fiveRandomArray.T)
print(f"The dot product of the array with itself is: \n{fiveRandomDotProduct}\n")
```

```python
'''
Problem 2

Description: Create two Numpy arrays and use them to visualize a sine wave using Matplotlib.
'''

# Module Imports
import numpy as np
import matplotlib.pyplot as plt

# Create X array with values from 0 to 100 incrementing by 0.1
sineWaveX = np.arange(0, 100, 0.1)

# Create Y array with y=sin(x)
sineWaveY = np.sin(sineWaveX)

# Plot the sine wave in matplotlib
plt.plot(sineWaveX, sineWaveY)

# Add x label, y label, and title
plt.xlabel("X Values")
plt.ylabel("Y Values (y=sin(x))")
plt.title("Sine Wave For x=0 to x=100")

# Save the plot as a PNG file
plt.savefig("sine_wave.png")
```

```python
'''
Problem 3

Description: Create two Numpy arrays and use them to visualize the quadratic function y = 0.5x^2 + 2x + 1 using Matplotlib.
'''

# Module Imports
import numpy as np
import matplotlib.pyplot as plt

# Create X array with values from 0 to 100 incrementing by 1
quadFunctionX = np.arange(0, 100, 1)

# Create Y array with y = 0.5x^2 + 2x + 1
quadFunctionY = (lambda x: (0.5 * (x * x)) + (2 * x) + 1)(quadFunctionX)

# Plot the function in matplotlib
plt.plot(quadFunctionX, quadFunctionY)

# Add x label, y label, title, and legend
plt.xlabel("X Values")
plt.ylabel("Y Values (y=0.5$x^2$ + 2x + 1)")
plt.title("Quadratic Function For x=0 to x=100")
plt.legend(['y = 0.5$x^2$ + 2x + 1'])

# Add grey gridlines with a dashdot line style
plt.grid(color='grey', linestyle='dashdot', linewidth=0.5)

# Save the plot as a PDF file
plt.savefig("quadratic_function.pdf")
```
